```python
import pandas as pd
```

```python
#1
df = pd.read_excel('sales_data.xlsx')
print(df.head())
```

```python
#2
total_sales_by_region = df.groupby('Region')['Sales'].sum().reset_index()
total_sales_by_region.columns = ['Region','Total_Sales']
print(total_sales_by_region)
```

```python
#3
highest_product_sales_by_region = df.groupby(['Region' , 'Product']) ['Sales'].sum().groupby('Region').idxmax()
print(highest_product_sales_by_region)
```

```python
#4
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df['Month'] = df['Date'].dt.month
```

```python
total_sales_by_month = df.groupby('Month')['Sales'].sum()
```

```python
month_with_highest_sales = total_sales_by_month.idxmax()
```

```python
print(f"Month with highest sales : {month_with_highest_sales}")
```
